```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
FECHA_DESDE = 2000
```

```python
df = pd.read_excel("https://thedocs.worldbank.org/en/doc/5d903e848db1d1b83e0ec8f744e55570-0350012021/related/CMO-Historical-Data-Monthly.xlsx",
                   sheet_name=1, header=4,index_col=0)
df = df.iloc[2:]
df.index = pd.to_datetime(df.index, format = "%YM%m")
df = df.replace('…', np.nan).replace('...', np.nan)
df = df.astype(float)
df = df[str(FECHA_DESDE):]
df
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter( x= df.index, y = df["Soybeans"], name = "Soybeans"))
fig.update_layout(template = None, font_family = "georgia")
```

```python
index_df = pd.read_excel("https://thedocs.worldbank.org/en/doc/5d903e848db1d1b83e0ec8f744e55570-0350012021/related/CMO-Historical-Data-Monthly.xlsx",
                   sheet_name=2, header=5,index_col=0)
index_df_columns_names = ["Total Index", "Energy", "Non-Energy", "Agriculture", "Beverages", "Food", "Oils & Meals", "Grains",
                          "Other Food", "Raw Materials", "Timber", "Other Raw Mat.",
                          "Fertilizers", "Metals & Minerals", 'Base Metals (ex. iron ore)',
                          "Precious Metals"]
index_df.index = pd.to_datetime(index_df.index, format = "%YM%m")
index_df = index_df.iloc[4:].astype(float)
index_df.columns = index_df_columns_names
index_df = index_df[str(FECHA_DESDE):]
index_df
```

```python
#Mostrar ciclos de commodities y eventos particulares
# eventos: guerra RUS-UCR(feb2022), Crisis 2008, Pandemia
#Ciclos
fig = go.Figure()
fig.add_trace(go.Scatter(x = index_df.index, y = index_df["Energy"], name = "Energía",line_width=3))
fig.add_trace(go.Scatter(x = index_df.index, y = index_df["Fertilizers"], name = "Fertilizantes",line_width=3))
fig.add_trace(go.Scatter(x = index_df.index, y = index_df["Agriculture"], name = "Agricultura",line_width=3))
fig.update_layout(template = None, title_text = "Evolución precios de grupos de materias primas seleccionadas", font_family = "georgia",
                  )
fig.add_annotation(text = "Elaboración propia. Fuente: Pink Sheet (Banco Mundial)", 
                   xref="paper", yref="paper", y=-.2, x = 0, showarrow=False)
fig.add_vline(x = "2012-01-01", line_dash = "dash",opacity=0.5)
fig.add_vline(x = "2004-01-01", line_dash = "dash",opacity=0.5)
fig.add_vline(x = "2021-01-01", line_dash = "dash",opacity=0.5)
fig.add_vline(x = "2022-02-01", line_dash = "dash",opacity=0.5, line_width =1)
fig.add_annotation(text="<b>Superciclo de commodities<br>\"Boom\" demanda china", x="2008", y = 275, showarrow=False)
fig.add_annotation(text="<b>Desaceleración china<br>Suba de la tasa de interés (FED)", x="2016-9", y = 275, showarrow=False)
fig.add_annotation(text = "<b>Guerra<br>Rusia-Ucrania", x="2022-02-01", showarrow=False, y = 320)
fig.add_vrect(x0="2020-02-01", x1="2021-01-01", row="all", col=1,
              annotation=dict(text = "COVID-19", textangle = 90,
                              align = "center",
                              font_size = 15,font_family ="georgia"),
              fillcolor="green", opacity=0.25, line_width=0)
fig.add_annotation(showarrow=False, text='@MartinBasualdo0', font=dict(size=16), font_family="arial", opacity=0.4,
                                    xref='paper', yref='paper', x=0.5, y=0.5)
```
